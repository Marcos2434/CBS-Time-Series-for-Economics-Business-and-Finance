τ → Reject when the test statistic is **more negative** than the critical value.  
φ → Reject when the test statistic is **greater** than the critical value.  
(τ tests for the unit root; φ tests for deterministic components — both are one-sided.)

---

### ADF Testing Workflow (φ-first logic)

#### 1. Start with specification: type = "trend"
- **φ₂**: joint test of constant and trend  
  - Reject → at least one (constant or trend) is significant → stay in "trend"  
  - Fail → both likely zero → skip drift → go directly to "none"
- **φ₃**: test of trend only  
  - Reject → trend is significant → keep "trend"  
  - Fail → drop trend → move to "drift"

Then check:
- **τ₃**: unit root test under trend specification  
  - Reject → series is **trend-stationary** (stationary after removing deterministic trend)  
  - Fail → series is **non-stationary (I(1))**

---

#### 2. If you move to specification: type = "drift"
- **φ₁**: test of constant (intercept)  
  - Reject → constant significant → keep "drift"  
  - Fail → drop constant → move to "none"

Then check:
- **τ₂**: unit root test under drift specification  
  - Reject → **mean-stationary** (stationary around non-zero mean)  
  - Fail → **non-stationary (I(1))**

---

#### 3. If you move to specification: type = "none"
- No φ test here.  
Then check:
- **τ₁**: unit root test without constant or trend  
  - Reject → **stationary around zero**  
  - Fail → **non-stationary (I(1))**

---

### Summary of interpretation
| Case | φ-test meaning | τ-test meaning | Stationarity type if τ rejects |
|------|----------------|----------------|--------------------------------|
| Trend | φ₂ (const+trend), φ₃ (trend) | τ₃ | Trend-stationary |
| Drift | φ₁ (const only) | τ₂ | Mean-stationary |
| None  | — | τ₁ | Stationary around zero |

---
